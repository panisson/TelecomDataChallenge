## Entropy

```python
intensity_data_ni = intensity_data[logical_not(intensity_data.Country_code.isin([0,39]))]
```

```python
def entropy(x):
    x = x.fillna(0)
    v = x[x > 0]
    v = v.div(float(v.sum()))
    v = v.map(lambda x: -x * log(x)).sum()
    return v
```

```python
def group_entropy(x):
    return x.pivot(index='Square_id', values='tot', columns='Country_code').fillna(0).apply(entropy, axis=1)
```

```python
daily_entropy = intensity_data_ni.groupby(['day']).apply(group_entropy)
```
